```python
import torch
import matplotlib.pyplot as plt
from matplotlib import ticker
```

```python
checkpoint = torch.load('C:/users/nickd/Documents/Visual_Code_Projects/PNN/checkpoints/pnn_model/pnn_model_2023_07_17-11_30_23.pth.tar')
```

```python
tr_metrics = checkpoint['tr_metrics']
val_metrics = checkpoint['val_metrics']
test_metrics = checkpoint['test_metrics']
```

```python
100000//50
```

```python
eval_steps = [100000, 200000, 300000, 400000,500000, 600000, 700000, 800000, 900000, 1000000]
formatter = ticker.ScalarFormatter()
formatter.set_scientific(False)
tr_range = [i * 50 for i in range(len(tr_metrics))[2000:]]

fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(15, 15))
ax[0,0].plot(tr_range, [metric['loss'] for metric in tr_metrics][2000:])
ax[0,0].set_title('Training loss')
ax[0,0].grid('on')
ax[0,0].set_xlabel('Epochs')
ax[0,0].xaxis.set_major_formatter(formatter)

ax[0,1].plot(eval_steps, [metric['loss'] for metric in val_metrics])
ax[0,1].set_title('Validation loss')
ax[0,1].xaxis.set_major_formatter(formatter)
ax[0,1].grid('on')
ax[0,1].set_xlabel('Epochs')

ax[0,2].plot(eval_steps, [metric['loss'] for metric in test_metrics])
ax[0,2].set_title('Test loss')
ax[0,2].xaxis.set_major_formatter(formatter)
ax[0,2].grid('on')
ax[0,2].set_xlabel('Epochs')

ax[1,0].plot(tr_range, [metric['psnr'] for metric in tr_metrics][2000:])
ax[1,0].set_title('Training PSNR')
ax[1,0].grid('on')
ax[1,0].set_ylabel('dB')
ax[1,0].set_xlabel('Epochs')
ax[1,0].xaxis.set_major_formatter(formatter)

ax[1,1].plot(eval_steps, [metric['psnr'] for metric in val_metrics])
ax[1,1].set_title('Validation PSNR')
ax[1,1].xaxis.set_major_formatter(formatter)
ax[1,1].grid('on')
ax[1,1].set_ylabel('dB')
ax[1,1].set_xlabel('Epochs')

ax[1,2].plot(eval_steps, [metric['psnr'] for metric in test_metrics])
ax[1,2].set_title('Test PSNR')
ax[1,2].xaxis.set_major_formatter(formatter)
ax[1,2].grid('on')
ax[1,2].set_ylabel('dB')
ax[1,2].set_xlabel('Epochs')

ax[2,0].plot(tr_range, [metric['ssim'] for metric in tr_metrics][2000:])
ax[2,0].set_title('Training SSIM')
ax[2,0].grid('on')
ax[2,0].set_xlabel('Epochs')
ax[2,0].xaxis.set_major_formatter(formatter)

ax[2,1].plot(eval_steps, [metric['ssim'] for metric in val_metrics])
ax[2,1].set_title('Validation SSIM')
ax[2,1].xaxis.set_major_formatter(formatter)
ax[2,1].grid('on')
ax[2,1].set_xlabel('Epochs')

ax[2,2].plot(eval_steps, [metric['ssim'] for metric in test_metrics])
ax[2,2].set_title('Test SSIM')
ax[2,2].xaxis.set_major_formatter(formatter)
ax[2,2].grid('on')
ax[2,2].set_xlabel('Epochs')

# Set common labels
```
